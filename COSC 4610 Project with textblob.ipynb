```python
#Use textblob to find neutral reviewers
#WInes given a high rating by a neutral reviewer is going to be liked by other people as well
from textblob import TextBlob

#Getting polarities of the wine descriptions the wine reviewers gave
polarities = []

for x in range(len(df)):
    sentence = TextBlob(df['description'][x])
    polarity = sentence.sentiment.polarity
    polarities = polarities + [polarity]
```

```python
#Add new column with the polarity of each wine description given by the wine reviewer
df['polarity'] = polarities
df.head()
```

```python
#The polarity score is a float within the range [-1.0, 1.0]
#Define a neutral review with a polarity within range [-2/3, 2/3]
neutralReviews = df.loc[(df['polarity'] <= (2/3)) & (df['polarity'] >= (-2/3))]

neutralReviews['taster_name'].value_counts()
#Disregard anonymous reviews
#Roger Voss, Michael Schachner, and Kerin O'Keefe are likely top 3 neutral reviewers
```

```python
neutralReviews.loc[neutralReviews['taster_name'] == 'Roger Voss']

#Get the wine reviews done by Roger Voss
Roger = df.loc[df['taster_name'] == 'Roger Voss']
print(max(Roger['points']))

#Get the wine reviews where Roger gave his highest rating 
#The characteristics of these wines could likely make a popular wine in the wine selling industry
Roger.loc[Roger['points'] == max(Roger['points'])]

```
